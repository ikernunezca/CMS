## Wilcoxon Tests (Supplementary Figure 5C): 
### Testing if genes associated to each disease tend to appear in the same communities in comparison to randomized versions of DisGeNET

```r
x <- 0
colnames(kazoo2)[ncol(kazoo2)] <- 'disease'
original <- kazoo2
original[,"disease"] <- as.character(original[,"disease"])
```

```r
head(original)
```

```r
Sys.time() # ~ 2.5 h run
```

```r
list_outs <- list()
for(i in 1:1000){
    random <- read.table(file= paste0("data/DisGeNET/random/RandomizationsWith12to30/AllDisgenetDiseasesGeneCommunityRandomizedAssociations_",i,".csv"),sep= ",",header= F,skip=1)
    colnames(random)[ncol(random)] <- 'disease'
    random <- as.matrix(random)
    random[,"disease"] <- as.character(random[,"disease"])
    rownames(random) <- random[,1]
    random <- random[,-1]

    zero <- c()
    for(j in 1:9){
        ####Random
          lst <- lapply(split(random[,j],random[,"disease"]),function(x) table(x))
          # freq <- lapply(lst, function(x) length(x[x>1])/length(unique(random[,"X1.5"])))
          freq <- lapply(lst, function(x) x)
          randomin <- unlist(freq)

        ###Original
          lst <- lapply(split(original[,j],original[,"disease"]),function(x) table(x))
          # freq <- lapply(lst, function(x) length(x[x>1])/length(unique(original[,"X1.5"])))
          freq <- lapply(lst, function(x) x)
          origin <- unlist(freq)
          res <- wilcox.test(randomin,origin)
          zero <- c(zero,res$p.value)
    }
message(print(Sys.time()))
list_outs[[i]] <- zero
}
```
